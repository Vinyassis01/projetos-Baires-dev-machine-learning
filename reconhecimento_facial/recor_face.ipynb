```python
from mtcnn import mtcnn
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
```

```python
img = plt.imread('/content/Captura de tela de 2025-08-22 16-09-00.png')
# Convert image to RGB if it's in RGBA format
if img.shape[-1] == 4:
    img = img[..., :3]
detector = mtcnn.MTCNN()
```

```python
def convert_to_rgb(image):
    if image.shape[-1] == 4:
        return image[..., :3]
    return image
```

```python
faces = detector.detect_faces(img)
```

```python
def imagens (img,faces):
    ax = plt.gca()
    for face in faces :
        x,y,width,height = face['box']
        rect = Rectangle((x,y),width,height,color='blue',fill=False,lw=4)
        ax.add_patch(rect)
    plt.imshow(img)
```

```python
rosto = imagens(img=img,faces=faces)
```
